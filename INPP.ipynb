```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mp = 938.272
mn = 939.565
me = 0.511
a_v = 15.67
a_s = 17.23
a_c = 0.714
a_a = 93.15
delta = 11.2
```

```python
def M(A,Z):
    N = A - Z
    mass = N*mn + Z*(mp + me) - a_v*A + a_s*A**(2/3) + a_c*Z**2/A**(1/3) + a_a*(N-Z)**2/(4*A)
    if N//2 == 0 and Z//2 == 0:
        mass -= delta/A**(1/2)
    elif N//2 != 0 and Z//2 != 0:
        mass += delta/A**(1/2)
    return mass
```

```python
M(235,92) - M(87,35) - M(145,57) - 3*mn
```
